```python
df['name']=df['name'].fillna('Not mentioned')
```

```python
df.dropna(subset='host_name',inplace=True)
```

```python
df['reviews_per_month']=df['reviews_per_month'].fillna(0)
```

```python
df['last_review']=df['last_review'].fillna(df['last_review'].mode()[0])
```

```python
df.drop(columns='neighbourhood_group',inplace=True)
```

```python
df['last_review']=pd.to_datetime(df['last_review'],dayfirst=True)
```

```python
df['name']=df['name'].str.replace(r'[^a-zA-Z0-9\s]', ' ', regex=True)
```
